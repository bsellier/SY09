```python
st = pd.read_csv("others_datasets/starbucks-menu-nutrition-drinks.csv")
st = st.rename(columns={'Unnamed: 0': 'product_name', 'Calories': 'calories', 'Fat (g)': 'total_fat_g',
                        'Fiber (g)': 'fiber_g', 'Carb. (g)': 'total_carbs_g', 'Sodium': 'sodium_mg'})
st.columns
```

```python
starbucks = pd.read_csv(
    "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-12-21/starbucks.csv"
)
starbucks.columns
```

```python
# Regarder si les datasets ont des produits en commun

def lower_list(lst):
    return [x.lower() for x in lst]

# Python program to illustrate the intersection
# of two lists in most simple way
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3

products_org = starbucks.product_name.unique().tolist()
#orgds = lower_list(orgds)

products_ds1 = st.product_name.unique().tolist()
#ds1 = lower_list(ds1)

intersection(products_ds1,products_org)
```

```python
# Vérifier qu'il ne s'agit pas des mêmes données

# Il faut qu'ils aient les mêmes colonnes

st.drop(columns=[x for x in st.columns if x not in starbucks.columns], inplace=True)
starbucks.drop(columns=[x for x in starbucks.columns if x not in st.columns], inplace=True)

# Unification des types
st = st.replace('-', np.nan)

for col in st.columns:
    if col == "product_name":
        continue
    st[col] = st[col].astype("float64")

    
starbucks = starbucks.replace('-', np.nan)

for col in starbucks.columns:
    if col == "product_name":
        continue
    starbucks[col] = starbucks[col].astype("float64")



# On retourne les lignes en communs

common_rows = st.merge(starbucks, how = 'inner' ,indicator=False)

print(f"Il y a {len(common_rows)} lignes en commun dans les deux datasets, ce qu'il faudra prendre en compte lors des prédictions")
```
